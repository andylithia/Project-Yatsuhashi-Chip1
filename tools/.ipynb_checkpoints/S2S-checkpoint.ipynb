```python
# spice to simulation

fdir  = '../mag/'
fname = 'PA_core_1'

fp = open(fdir+fname+'.spice','r')
lines = fp.readlines()
fp.close()
```

```python
namelist = []
obuf = []
in_body = False
starting_point = None
for i in range(len(lines)):
    line = lines[i]
    if line.startswith('.subckt '+fname):
        print('Starting @'+str(i))
        starting_point = i
        in_body = True
    else:
        if(in_body):
            parts = line.split(' ')[1:-1]
            # WARNING: Not dealing with + lines
            for j in parts:
                if j not in namelist:
                    if j!='0':
                        namelist.append(j)
                        print('Port Added: '+j)
        else:
            if(line[0]!='*') and (len(line)>1):
                obuf.append('    '+line)
    
tagline = '.subckt '+fname
for e in namelist:
    tagline = tagline+' '+e
obuf = [tagline+'\n']+obuf
for i in range(starting_point+1, len(lines)):
    # Append subckt contents
    obuf.append(lines[i])
    
fout = open(fdir+fname+'.sim.spice','w+')
fout.writelines(obuf)
fout.close()
        
```
